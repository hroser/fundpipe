```
import random

funds = []
refunds = []
ownerrefunds = []
total_volume = 0
total_bonus = 0
total_funding = 0
total_fee = 0
total_payback_close = 0
funding_rate = 0.5
funding_fee = 0.04
inflation_rate = 0.005
fundmax = 10
last_profit = 0

for i in range(1000):
  fund = random.randint(1, int(fundmax))
  for i in range(len(refunds)):
    refunds[i] += ((1 - funding_rate) * fund) * (float(funds[i]) / sum(funds))
  funds.append(fund)
  refunds.append(0)
  ownerrefunds.append(0)
  fundmax = (1 + inflation_rate)*fundmax
  
for i in range(len(funds)):
  total_volume = total_volume + funds[i]
  total_bonus = total_bonus + max(0, refunds[i] - funds[i])
  total_funding = total_funding +  funding_rate * (1 - funding_fee) * funds[i]
  total_fee = total_fee +  funding_rate * funding_fee * funds[i]
  total_payback_close = total_payback_close +  max(0, funds[i] - refunds[i])
  if (i+1 < len(funds)) and (refunds[i] + ownerrefunds[i] - funds[i] > 0) and (refunds[i+1] + ownerrefunds[i+1] - funds[i+1]):
    last_profit = i

needed_to_close = funds[len(funds)-1]/(float(funds[len(funds)-1]) / sum(funds))
    
print("total volume {:.2f}".format(total_volume))
print("total funding {:.2f}".format(total_funding))
print("total fee (paid to fundpipe) {:.2f}".format(total_fee))
print("total bonus (paid to funders) {:.2f}".format(total_bonus))
print("total payback for closing {:.2f}, percent of total funding {:.2f}".format(total_payback_close, 100.0*total_payback_close/total_funding))
print("\n")
print("last profit at {}".format(last_profit))
print("needed to close {}".format(needed_to_close))
print("\n")

for i in range(len(funds)):
  print("{}: funded {:.2f}, refunded {:.2f}, bonus {:.2f}, balance: {:.2f}, bonus percentage: {:.2f}".format(i, funds[i], ownerrefunds[i], refunds[i], refunds[i] + ownerrefunds[i] - funds[i], (100.0 * (refunds[i] + ownerrefunds[i] - funds[i])) / funds[i] ))
```
